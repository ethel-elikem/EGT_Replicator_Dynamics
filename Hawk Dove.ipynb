```python
def Hawk_dove():
    v= int(input("Enter resource: "))
    c= int(input("Enter cost: "))
    d= int(input("Enter number of doves: "))
    h= int(input("Enter number of hawks: "))
    g= int(input("How many generation: "))

    n = d+h         #Total population
    Xd= d/n         #Proportion of doves
    Xh= h/n         #Proportion of hawks
    darr=[Xd]
    harr=[Xh]

    for t in range(g):
        fd = (Xd*(v/2)) + (Xh*0)             #Payoff for playing dove strategy
        fh = (Xh*((v-c)/2)) + (Xd*v)         #Payoff for plating hawk strategy
        f = (Xd*fd) + (Xh*fh)                #Average fitness
        dchange = Xd*(fd-f)                  #Change in dove proportion
        hchange = Xh*(fh-f)                  #Change inn hawk proportion

        Xd += dchange
        Xh += hchange
        darr.append(Xd)
        harr.append(Xh)

    plot(darr, label = 'doves')
    plot(harr, label = 'hawks')
    grid()
    title("Dove proportion VS Hawk proportion")
    legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2)

```
